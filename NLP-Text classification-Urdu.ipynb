```python
import os
import re
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
import pandas as pd
data = []
# Folder with .txt files
folder_path = 'khazeena Urdu-NER master Corpus/'
for fname in os.listdir(folder_path):
    if fname.endswith('.txt'):
        file_path = os.path.join(folder_path, fname)
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read().strip()
            data.append({'filename': fname, 'text': text})

df = pd.DataFrame(data)
df.to_csv('merged_urdu_text.csv', index=False, encoding='utf-8')
```

```python
import urduhack

urduhack.download()
from urduhack.normalization import normalize
from urduhack.tokenization import sentence_tokenizer, word_tokenizer

text = open("merged_urdu_text.csv", encoding="utf-8").read()
text = normalize(text)
sentences = sentence_tokenizer(text)

print(f"Total sentences: {len(sentences)}")
```

```python
sentences[:5]
```

```python
pip install gensim
```

```python
from gensim.models import Word2Vec

# Train model on tokenized sentences
model = Word2Vec(
    sentences=sentences,         # list of tokenized sentences
    vector_size=100,          # embedding dimensions
    window=5,                 # context window
    min_count=2,              # ignore rare words
    workers=4,                # number of CPU threads
    sg=1                      # 1=skip-gram, 0=CBOW
)

# Save the model (optional)
model.save("urdu_word2vec.model")
```
